# seaborn概述

Seaborn是一个基于matplotlib的Python数据可视化库。它提供了一个高级接口，用于绘制有吸引力且信息丰富的统计图形。

网址：http://seaborn.pydata.org/



Seaborn的一些主要特点和功能：
- 内置统计图表：Seaborn提供了许多常见的统计图表，如柱状图、直方图、箱线图、核密度图、散点图等。这些图表可以根据不同变量或类别自动进行分组和汇总。
- 数据探索和可视化：Seaborn拥有丰富的数据探索工具，包括通过展示变量之间的关联、分布和差异来揭示数据的模式。用户可以轻松地查看多个变量之间的关系，并进行相应的分析。
- 绘图风格和颜色：Seaborn提供了多种内置的绘图风格和配色方案，可以快速设置图表的外观。这些风格和颜色方案具有专业设计感，使生成的图表具有美观的外观。
- 数据集可视化：Seaborn还提供了几个内置数据集，用于演示和实践数据可视化。这些数据集可用于绘制具有现实意义的图表，并帮助用户更好地理解Seaborn的功能和使用方法。
- 与Pandas集成：Seaborn非常适合与Pandas库一起使用。它可以直接从Pandas数据框导入数据，并支持对数据进行灵活的操作和可视化。

# seaborn库的介绍和使用

Seaborn库主要分为以下几个模块：
- 绘图接口（Plotting Interface）：这是Seaborn库的核心部分，提供了大量用于绘制各种统计图形的函数。例如，relplot()函数用于绘制关系图，distplot()函数用于绘制分布图，boxplot()函数用于绘制箱线图等。
- 风格接口（Styling Interface）：这个接口允许用户自定义Seaborn图表的样式，包括颜色、字体、线条样式等。例如，set_style()函数用于设置图表的整体风格，set_palette()函数用于设置颜色方案等。
- 调色板（Color Palettes）：Seaborn库提供了一系列预设的调色板，用户可以直接使用这些调色板为图表设置颜色。调色板可以根据数据的性质（如连续变量、分类变量等）进行选择。
- 统计模型可视化（Statistical Model Visualization）：这个模块提供了用于可视化统计模型的函数，如回归线、置信区间等。例如，lmplot()函数用于绘制线性回归图，regplot()函数用于绘制回归线图等。
- 数据集（Datasets）：Seaborn库还内置了一些用于示例的数据集，方便用户学习和使用。

## 实战演练

```python
#实战一  画直方图
import seaborn as sns

penguins = sns.load_dataset("penguins")  #加载自带的数据集

#调用hisplot画出直方图
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
#实战二  画核密度估计图
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
# 实战三-- 散点图
from matplotlib import pyplot as plt  #配合matplotlib一起使用

f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3])) #显示2个图

#通过调用scatterplot来画散点图
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0]) 
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```

```python
#实战四--
tips = sns.load_dataset("tips")  #加载数据

#画relplot图 关系图
g = sns.relplot(data=tips, x="total_bill", y="tip")
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5, 2))
```

```python
# 实战5-- jointplot() 函数绘制了两个变量之间的关系或联合分布，同时添加了边缘轴，用于分别显示每个变量的单变量分布
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
#实战6--pairplot和jointplot类似 也是结合了联合视图（joint views）和边缘视图（marginal views）。但是，pairplot() 并不专注于单一的关系，而是同时可视化变量之间的每一对组合
sns.pairplot(data=penguins, hue="species")
```

```python
#箱形图
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="box")  #通过kind=box来定义形状
```

```python
#violin图
sns.catplot(data=tips, x="total_bill", y="day", hue="sex", kind="violin")  #通过violin来定义形状
```

```python
#通过kind=bar来定义形状
titanic = sns.load_dataset("titanic")
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="bar")
```

```python
#展示额外的额度信息 catplot() 建立在 FacetGrid 的基础上，很容易添加分组变量来可视化更高维度的关系
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker",kind="swarm", col="time", aspect=.7)
```
